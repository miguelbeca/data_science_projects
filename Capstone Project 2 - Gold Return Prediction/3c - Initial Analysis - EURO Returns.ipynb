```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
#Read data

data = pd.read_csv('data/analysis/D1_EURUSD_Clean.csv', parse_dates=[0])
data.set_index('Date',inplace=True)
print(data.head())
```

```python
#Plot timeseries
fig, ax = plt.subplots(figsize=(16,8));
data['Close_EUR'].plot(ax=ax);
ax.set_title('Euro daily price, 2003-2019');
ax.set_ylabel('USD($)');
ax.set_xlabel('Year');
ax.xaxis.set_ticks_position('bottom')
fig.tight_layout();
```

```python
#Returns data
data['Daily_return_EUR'] = data.Close_EUR.pct_change().mul(100)
print(data.tail())
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

data['Daily_return_EUR'].plot(ax=ax);

ax.set_title('Euro daily returns, 2003-2019 (%)');
ax.xaxis.set_ticks_position('bottom')
fig.tight_layout();
```

```python
#Using log returns
data['Log_daily_Return_EUR'] = data['Close_EUR'].apply(lambda x: np.log(x)).diff()
print(data.head())
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

data['Log_daily_Return_EUR'].plot(ax=ax);

ax.set_title('Euro log daily returns, 2003-2019 (%)');
ax.xaxis.set_ticks_position('bottom')
fig.tight_layout();
```

```python
#Save data with returns for future analysis
data.to_csv('data/Analysis/D1_EURUSD_Returns.csv')
```
