```python
import pandas as pd
import numpy as np
import datetime
import talib as ta
import matplotlib.pyplot as plt
import seaborn as sns
import talib as ta
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read data
data = pd.read_csv('data/analysis/D1_XAUUSD_Returns.csv', parse_dates=[0], index_col='Date')
print(data.head())
```

```python
#Analysis of daily returns by Weekday and Day of the Year
daily_ret = data.iloc[:,[7,5]].dropna()
daily_ret.head()
```

```python
#Daily returns by weekday count, mean and quartiles
daily_ret.groupby('Weekday').describe().loc[['Sunday','Monday','Tuesday','Wednesday', 'Thursday', 'Friday']]
```

```python
#Plotting distribution of returns
daily_ret_Mon = daily_ret[daily_ret['Weekday']== 'Monday']
daily_ret_Tue = daily_ret[daily_ret['Weekday']== 'Tuesday']
daily_ret_Wed = daily_ret[daily_ret['Weekday']== 'Wednesday']
daily_ret_Thu = daily_ret[daily_ret['Weekday']== 'Thursday']
daily_ret_Fri = daily_ret[daily_ret['Weekday']== 'Friday']
```

```python
#kwargs = dict(bins=30) **kwargs
fig, axes = plt.subplots(nrows=5, ncols=1, sharex=True, sharey=True, figsize=(8,16))

sns.distplot(daily_ret_Mon.Daily_return,ax=axes[0]).set_title('Monday')
axes[0].set_xlabel('')
sns.distplot(daily_ret_Tue.Daily_return, ax=axes[1]).set_title('Tuesday')
axes[1].set_xlabel('')
sns.distplot(daily_ret_Wed.Daily_return, ax=axes[2]).set_title('Wednesday')
axes[2].set_xlabel('')
sns.distplot(daily_ret_Thu.Daily_return, ax=axes[3]).set_title('Thursday')
axes[3].set_xlabel('')
sns.distplot(daily_ret_Fri.Daily_return, ax=axes[4]).set_title('Friday')
axes[4].set_xlabel('')
fig.tight_layout()

#plt.show()
```

```python
#Exclude Sunday data & print violin plot
data_ex = data[data['Weekday']!= 'Sunday']
fig,ax = plt.subplots(figsize=(20,15))
sns.violinplot(ax=ax, x="Weekday", y="Daily_return", data = data_ex)
```
